```python
# import pandas to create DataFrame
import pandas as pd

#make DataFrame of the given data
data = pd.DataFrame({"A":[1,2,4,1,2,4],
                    "B":[4,5,6,7,8,9],
                    "c":[-0.2,0,0,0,0,0],
                    "D":[1,1,1,1,1,1]})
```

```python
data.var()
```

```python
### it will check for cols with var 2.0
from sklearn.feature_selection import VarianceThreshold
var_thresh = VarianceThreshold(threshold=2.0)
var_thresh.fit(data)
```

```python
VarianceThreshold(threshold=2.0)
```

```python
var_thresh.get_support()
```

```python
data.columns[var_thresh.get_support()]
```

```python
constant_columns = [column for column in data.columns
                    if column not in data.columns[var_thresh.get_support()]]
print (len(constant_columns))
```

```python
for feature in constant_columns:
    print(feature)
```

```python
data.drop(constant_columns, axis=1, inplace=True)
```
